```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("/kaggle/input/analysts-is/cleaned_ready_dataset_Pranav_PRB9282.csv")
data['return_on_assets'] = data['ni_mean'] / data['at_mean']

plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='gsector_mean_y', y='return_on_assets', palette='Set3')
plt.title("Return on Assets (ROA) by Sector")
plt.xlabel("Sector")
plt.ylabel("ROA")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("/kaggle/input/analysts-is/cleaned_ready_dataset_Pranav_PRB9282.csv")

# Calculate operating margin
data['operating_margin'] = (data['sale_mean'] - data['cogs_mean'] - data['xsga_mean']) / data['sale_mean']

# Group by sector and calculate mean operating margin
sector_op_margin = data.groupby('gsector_mean_y')['operating_margin'].mean()

# Plotting as a bar chart to handle possible negative values
plt.figure(figsize=(10, 6))
sns.barplot(x=sector_op_margin.index, y=sector_op_margin.values, palette="pastel")
plt.title("Average Operating Margin by Sector")
plt.xlabel("Sector")
plt.ylabel("Average Operating Margin")
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load and prepare your dataset
data = pd.read_csv("/kaggle/input/analysts-is/cleaned_ready_dataset_Pranav_PRB9282.csv")
data.replace([float('inf'), -float('inf')], pd.NA, inplace=True)

# Calculate the Capex to Sales ratio
data['capex_to_sales'] = data['ppent_mean'] / data['sale_mean']

# Plot Capex to Sales by Sector using Stripplot
plt.figure(figsize=(10, 6))
sns.stripplot(data=data, x='gsector_mean_y', y='capex_to_sales', jitter=True)
plt.title("Capital Expenditures to Sales Ratio by Sector")
plt.xlabel("Sector")
plt.ylabel("Capex to Sales")
plt.xticks(rotation=45)
plt.show()
```
